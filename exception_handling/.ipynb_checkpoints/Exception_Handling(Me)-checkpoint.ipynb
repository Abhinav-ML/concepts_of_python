# Q:Why to write class name with except?
* Because by mentioning the class name we are able to understand which type of error we are getting.

```python
a = int(input("Enter 1st No:"))
b = int(input("Enter 2nd No:"))
try:
    c = a/b
    print(c)
    d = a + str(b)
    print(d)
except ZeroDivisionError:
    print("b should not be zero")
```

```python
a = int(input("Enter 1st No:"))
b = int(input("Enter 2nd No:"))
try:
    c = a/b
    print(c)
    d = a + str(b)
    print(d)
except ZeroDivisionError:
    print("b should not be zero")
except:
    print('Something went wrong')
# here we are unable to find what type of error is in our program, the good thing is we handle the exception
```

```python
a = int(input("Enter 1st No:"))
b = int(input("Enter 2nd No:"))
try:
    c = a/b
    print(c)
    d = a + str(b)
    print(d)
except TypeError: #(TypeError should be before of ZeroDivisionError bcoz of exception class hierarchy )
    print("value should be integer only")
except ZeroDivisionError:
    print("b should not be zero")
except:
    print('Something went wrong')
# here we can easily find the type of error TypeError by viewing the print statement and output statement
```

```python
a = int(input("Enter 1st No:"))
b = int(input("Enter 2nd No:"))
try:
    c = a/b
    print(c)
    d = a + str(b)
    print(d)
except (ZeroDivisionError,TypeError): # another way to write multiple type of error 
    print("b should not be zero or value should be integer only") # common message for both typr of error
except:
    print('Something went wrong')
```

## exc_info()

```python
import sys
while True:
    try:
        a=int(input("first number:"))
        b=int(input("second number:"))
        c=a/b
        print("div:", c)
        break
    except:
        a,b,c=sys.exc_info()
        print("exception class", a)
        print("exception message", b)
        print("line number", c.tb_lineno)
```

## format_exc()

```python
import traceback
while True:
    try:
        a=int(input("first number:"))
        b=int(input("second number:"))
        c=a/b
        print("div:", c)
        break
    except:
        print(traceback.format_exc())
```

## Custom Exception
exception class in python already they have define and we are calling this class.
if you are going to define this class in your code so it will run

```python
class NegativeNumberException(Exception):
    pass
```

```python
while True:
    try:
        a = int(input("first no:"))
        b = int(input("second no:"))
        if a<0 or b<0:
            raise NegativeNumberException ("-ve no. not allowed")
        c = a/b
        print(c)
        break
    except NegativeNumberException as e:
        print(e)
```

```python
#custom exception
class NegativeNumberException(Exception):
    pass
while True:
    try:
        a=int(input("first number:"))
        b=int(input("second number:"))
        if a<0 or b<0: 
            raise NegativeNumberException("neg number note allowd")
        c=a/b
        print("div is", c)
        break
    except ValueError:
        print("please enter int only")
    except ZeroDivisionError:
        print("please enter non zero denomitor")
    except NegativeNumberException as e:
        print(e)
```
